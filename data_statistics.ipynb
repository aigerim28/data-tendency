## Percentiles
____
**Percentiles** are used in statistics to give you a number that describes the value that a given percent of the values are lower than.

To find the percentile we use **np.percentile(column_variable, percent)** method:

```python
# Create the variable for column to isolate it from other columns of DataSet.
population = cal_house_df['population']

# Use the function for percentile.
percentile25 = np.percentile(population, 25)

# Display it on the screen.
print('The 25 percentile of population is:', percentile25)
```

##### Now, we can conclude that 75% of population values are greater than 780 people.

Let's find another percentile value for this column.

```python
# Use the function for percentile.
percentile10 = np.percentile(population, 10)

# Display it on the screen.
print('The 10 percentile of population is:', percentile10)
```

##### We can see that the 90% of the population is more than 498 people, and 10% is less than 498 people. 

## Standard Deviation
____
**Standard deviation** is a number that describes how spread out the observations are.
- A mathematical function will have difficulties in predicting precise values, if the observations are "spread". Standard deviation is a measure of uncertainty.

- A low standard deviation means that most of the numbers are close to the mean (average) value.

- A high standard deviation means that the values are spread out over a wider range.

```python
# We can use the std() function from Numpy to find the standard deviation of a variable.
std = np.std(cal_house_df)

# Display on the screen.
std
```

## Coefficient of Variation
_____
- **The coefficient of variation** is used to get an idea of how large the standard deviation is.

- Mathematically, the coefficient of variation is defined as:

  **Coefficient of Variation = Standard Deviation / Mean**

We can do this in Python if we proceed with the following code:

```python
# Create a variable using the Numpy functions: std() and mean()
cv = np.std(cal_house_df) / np.mean(cal_house_df)

# Display on the screen.
cv
```

## Variance
_____
- Variance is another number that indicates how spread out the values are.

In fact, if you take the square root of the variance, you get the standard deviation. Or the other way around, if you square the standard deviation we can get the variance.

```python
# Calculate the variation with np.var() method
var = np.var(cal_house_df)

# Display on the screen
var
```

## Correlation
_____
**Correlation** measures the relationship between two variables.

We mentioned that a function has a purpose to predict a value, by converting input (x) to output (f(x)). We can say also say that a function uses the relationship between two variables for prediction.

### Correlation Coefficient
**The correlation coefficient** measures the relationship between two variables.

The correlation coefficient can never be less than -1 or higher than 1.

- 1 = there is a perfect linear relationship between the variables (if one of the variables increases, the second increases as well)
- 0 = there is no linear relationship between the variables
- -1 = there is a perfect negative linear relationship between the variables (if one increases, the other one decreases)

```python
print('Example of positive correlation coefficient. ')

# Create a plot to show the relation between two variables.
cal_house_df.plot(x='total_rooms', y='total_bedrooms', kind='scatter')

# Display the plot
plt.show()
```

Here we can conclude that as the number of total_rooms increases the number of the total_bedrooms also increases.

```python
print('Example of negative correlation coefficient. ')

# Create a plot to show the relation between two variables.
cal_house_df.plot(x = 'longitude', y='latitude', kind='scatter')

# Display the plot
plt.show()
```

```python
print('Example of 0 correlation coefficient. ')

# Create a plot to show the absence of relation between two variables.
cal_house_df.plot(x='population', y='median_house_value', kind='scatter')

# Display the plot
plt.show()
```

### Correlation Matrix
_____
A matrix is an array of numbers arranged in rows and columns.

A **correlation** matrix is simply a table showing the correlation coefficients between variables.

```python
# Create a correlation matrix as a table with the corr() function
corr_matrix = round(cal_house_df.corr(), 2)

# Display the correlation matrix
print(corr_matrix)
```

We can also display the correlation matrix with the help of **Heatmap.**

```python
# Create a new DataFrame for correlation matrix
cal_h_corr = cal_house_df.corr()

# Create a heatmap from the correlaiton matrix using sns.heatmap() function
hp_corr = sns.heatmap(cal_h_corr,
vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(50, 500, n=500),
square=True
)

# Display on the screen. 
plt.show()
```

## Causality VS Correlation
_____
**Causality** is th relationship between two variables, where x is the cause of y. It seems similar to correlation, but there is a big difference between them:

* Correlation is a number that measures how closely the data are related
* Causality is the conclusion that x causes y.

So, any variables can be related with each other accidentally and don't show any relation between them. 

#### Example
Let us assume that we have some data about the number of students coming to some market and the sales of socks sold that day. We know that the students did not buy any socks, but it seems like there is some correlation between them accidentally. 

```python
# Create the dictionary and the DataFrame from it
sales_stud_df = {'students_number': [10, 11, 18, 20, 22, 24, 30, 34, 40, 56, 60, 65],
      'sales_of_socks': [8, 12, 16, 20, 24, 28, 32, 36, 40, 44, 48, 52]}
sales_stud_df = pd.DataFrame(data= sales_stud_df)

# Create and display the correlation plot.
sales_stud_df.plot(x='students_number', y='sales_of_socks', kind='scatter')
plt.show()

# The correlation matrix between two variables.
corr_market = sales_stud_df.corr()
print('The matrix of correlation coefficients:')
print()
print(corr_market)
```

##### Here we can see that the correlation between the number of students and sales of socks  is **positive.** 


But can we say that the number of students coming to the market is the cause of the big sales of socks?
**Definitely, not.** 

The possible causes of the big sales of socks:
- Cold weather 
- The 23rd of February 
- The difficulties with the Heating system of the city
- New collection of socks with original design
- Big discounts etc. 

### Hence, 
we can not say that there is the relation between two variables by only seeing the positive correlation coefficient. We must conclude the relationships and causality using **the common sense and the logical connections.**
